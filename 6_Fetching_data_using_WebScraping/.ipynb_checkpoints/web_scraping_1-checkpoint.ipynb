```python
import pandas
import requests
from bs4 import BeautifulSoup
```

```python
base_url = 'http://books.toscrape.com/'
response = requests.get(base_url)
response.encoding = 'utf-8' # Force correct encoding for better output
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup.find_all('h1')[0].text
```

```python
titles = []
for h3 in soup.find_all('h3'):
    full_title = h3.find('a')['title']
    titles.append(full_title)
```

```python
for title in titles:
    print(title)
```

```python
soup.find_all('p')
```

```python
ratings = []

for article in soup.find_all('article', class_='product_pod'):
    p_tag = article.find('p', class_='star-rating')
    # The rating is the second class of the <p> tag
    rating = p_tag['class'][1]
    ratings.append(rating)
```

```python
#print ratings
for rate in ratings:
    print(rate)
```

```python
prices = []

for div in soup.find_all('div', class_='product_price'):
    p_tag = div.find('p', class_='price_color')
    prices.append(p_tag.text.strip())
```

```python
for price in prices:
    print(price)
```
