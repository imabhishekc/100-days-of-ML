```python
import requests

url = "https://api.themoviedb.org/3/movie/top_rated?language=en-US&page=1"

headers = {
    "accept": "application/json",
    "Authorization": "Bearer YOUR_TOKEN_HERE"
}

response = requests.get(url, headers=headers)

df = pd.DataFrame(response.json()['results'])[['id', 'title', 'original_language', 'overview', 'popularity', 'release_date', 'vote_average', 'vote_count']]
```

```python
df = pd.DataFrame()
df
```

```python
for i in range(1, 401):
    url = f"https://api.themoviedb.org/3/movie/top_rated?language=en-US&page={i}"
    headers = {
        "accept": "application/json",
        "Authorization": "Bearer YOUR_TOKEN_HERE"
    }
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        data = response.json().get('results', [])
        if data:
            temp_df = pd.DataFrame(data)[['id', 'title', 'original_language', 'overview', 'popularity', 'release_date', 'vote_average', 'vote_count']]
            df = pd.concat([df, temp_df], ignore_index=True)
    else:
        print(f"Failed to fetch page {i}: Status Code {response.status_code}")
```
