```python
data
```

```python
station_id = data.get('id', 'unknown_station')
```

```python
hourly = data.get('hourlyMeasurements', [])
daily = data.get('dailyMeasurements', [])
```

```python
df_hourly = pd.DataFrame(hourly)
df_daily = pd.DataFrame(daily)
```

```python
df_hourly['station_id'] = station_id
df_daily['station_id'] = station_id
```

```python
df_hourly.to_csv("hourly_measurements.csv", index=False)
df_daily.to_csv("daily_measurements.csv", index=False)
```
